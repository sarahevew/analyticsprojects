```python
import pandas as pd
```

```python
data = pd.read_csv('/Users/sarahweinflash/Desktop/Education/Projects/New_York_City_Leading_Causes_of_Death.csv')
```

```python
data
```

```python
Gender = []
for i in data['Sex']:
    if i == 'M':
        Gender.append('Male')
    elif i == 'F':
        Gender.append('Female')
    elif i == 'Male':
        Gender.append('Male')
    elif i == 'Female':
        Gender.append('Female')

Race = []
for i in data['Race Ethnicity']:
    if i == 'White Non-Hispanic':
        Race.append('White')
    elif i == 'Hispanic':
        Race.append('Hispanic')
    elif i == 'Not Stated/Unknown':
        Race.append('Unknown')
    elif i == 'Asian and Pacific Islander':
        Race.append('AAPI')
    elif i == 'Black Non-Hispanic':
        Race.append('Black')
    elif i == 'Other Race/ Ethnicity':
        Race.append('Other')
    elif i == 'Non-Hispanic Black':
        Race.append('Black')
    elif i == 'Non-Hispanic White':
        Race.append('White')

Cause = []
for i in data['Leading Cause']:
    if "Respiratory" in i: 
        Cause.append('Respiratory')
    elif "Cancer" in i:
        Cause.append('Cancer')
    elif "Cirrhosis" in i:
        Cause.append('Cirrhosis')
    elif "Influenza" in i:
        Cause.append('Influenza')
    elif "Accidents" in i:
        Cause.append('Accident')
    elif "Renal" in i:
        Cause.append('Renal Disease')
    elif "Stroke" in i:
        Cause.append('Stroke')
    elif "HIV" in i:
        Cause.append('HIV')
    elif 'Other' in i:
        Cause.append('Other')
    elif 'Heart' in i:
        Cause.append('Heart Disease')
    elif 'Mental' in i:
        Cause.append('Mental Disease')
    elif 'Homicide' in i:
        Cause.append('Homicide')
    elif 'Aneurysm' in i:
        Cause.append('Aneurysm')
    elif 'Diabetes' in i:
        Cause.append('Diabetes')
    elif 'Septicemia' in i:
        Cause.append('Blood Poisoning')
    elif 'Congenital' in i:
        Cause.append('Congenital Disease')
    elif 'Hepatitis' in i:
        Cause.append('Hepatitis')
    elif 'Nephrisis' in i:
        Cause.append('Renal Disease')
    elif 'Tuberculosis' in i:
        Cause.append('Tuberculosis')
    elif 'Suicide' in i:
        Cause.append('Suicide')
    elif "Alzheimer's" in i:
        Cause.append("Alzheimers")
    elif 'Neoplasms' in i:
        Cause.append('Cancer')
    elif 'Alcohol' in i:
        Cause.append('Alcohol Poisoning')
    elif "Parkinson's" in i:
        Cause.append('Parkinsons')
    elif 'Atherosclerosis' in i:
        Cause.append('Heart Disease')
    elif 'Peptic' in i:
        Cause.append('Ulcer')
    elif 'Complications' in i:
        Cause.append('Medical Complications')
    elif 'Cholelithiasis' in i:
        Cause.append('Galbladder Disease')
    elif 'Anemias' in i:
        Cause.append('Anemia')
    elif 'Meningitis' in i:
        Cause.append('Meningitis')
    elif 'Pregnancy' in i:
        Cause.append('Pregnancy & Childbirth')
    else:
        Cause.append('Other')
```

```python
data['Gender'] = Gender
data['Race'] = Race
data['Cause'] = Cause
data = data[['Year','Cause','Gender','Race','Deaths']]
data.to_csv('Causes_of_Death.csv')
```
