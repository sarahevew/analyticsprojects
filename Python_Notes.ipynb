## Logic, Control Flow, & Filtering

```python
# Less Than: <
# Greater Than: >
# Less Than or Equal To: <=
# Greater Than or Equal To: >=
# Equal To: ==
# Not Equal To: !=

# and
True and True # Result: True
(10 > 5) and (4 < 5) # Result: True
True and False # Result: False
(10 > 5) and (6 < 5) # Result: False
False and False # Result: False
(3 > 5) and (6 < 5) # Result: False

# or
True or True # Result: True
(10 > 5) or (4 < 5) # Result: True
False or True # Result: True
(10 > 5) or (6 < 5) # Result: True
False or False # Result: False
(3 > 5) or (6 < 5) # Result: False

# not
not True # Result: False
not (10 > 5) # Result: False
not False # Result: True
not (6 < 5) # Result: True

# np.logical
import numpy as np
a = np.array([8,6,7,5,3,0,9])
a < 3 # Result: array([False, False, False, False, False,  True, False])
a[a<7] # Result: array([6, 5, 3, 0])
np.logical_and(a<7, a>3) # Result: array([False,  True, False,  True, False, False, False])
a[np.logical_and(a<7, a>3)] # Result: array([6, 5])
np.logical_or(a<4, a>8) # Result: array([False, False, False, False,  True,  True,  True])
a[np.logical_or(a<4, a>8)] # Result: array([3, 0, 9])
np.logical_not(a<8, a>5) # Result: array([ True, False, False, False, False, False,  True])
a[np.logical_not(a<8, a>5)] # Result: array([8, 9])

# np.logical - DataFrames
stock = {'apples':[5,9,2], 'bananas':[9,2,4], 'carrots':[8,3,4]}
stock = pd.DataFrame(stock)
np.logical_and(stock['apples']>4, stock['bananas']<5) # Result: False, True, False; Series
stock[np.logical_and(stock['apples']>4, stock['bananas']<5)] # Result: 9, 2, 3; DataFrame

# if/elif/else
z = 9
if z % 2 == 0:
    print('z is divisible by 2')
elif z % 3 == 0:
    print('z is divisible by 3, but not 2')
else:
    print('z is not divisible by 2 or 3')
```

## Loops

```python
# While Loop
error = 3
print('While Loop')
while error > 0:
    print('Error has occurred... recalibrating')
    error = error - 1
print('The program is error free!','\n')

# For Loop - list
print('For Loop - List')
x = [9,3,5,7,3]
for i in x:
    print('The number is', i)
print('\n')

# For Loop - list with index
print('For Loop - List, with Indices')
x = [9,3,5,7,3]
for index, value in enumerate(x):
    print('The number at index ', index, ' is ', value)
print('\n')

# For Loop - string
print('For Loop - String')
y = 'Amy'
for i in y:
    print(i.capitalize())
print('\n')

# For Loop - dictionary
print('For Loop - Dictionary')
stock = {'apples':[5,9,2], 'bananas':[9,2,4], 'carrots':[8,3,4]}
for key, value in stock.items():
    print('The product is:', key, 'and the inventories are:', value)
print('\n')

# For Loop - dictionary
print('For Loop - Dictionary, Decomposed')
for key, value in stock.items():
    for i in value:
        print('This store has', i, key, 'in stock')
print('\n')

# For Loop - numpy array
print('For Loop - Numpy Array')
shirts = np.array([[3,4,5],[9,4,2],[5,5,6]])
for i in shirts:
    print(i)
print('\n')

# For Loop - numpy array with index
print('For Loop - Numpy Array, Decomposed')
shirts = np.array([[3,4,5],[9,4,2],[5,5,6]])
for i in np.nditer(shirts):
    print(i)
print('\n')

# For Loop - DataFrame
stock = {'apples':[5,9,2], 'bananas':[9,2,4], 'carrots':[8,3,4]}
stock = pd.DataFrame(stock)
stock.index = ['Store 1', 'Store 2', 'Store 3']
print('For Loop - DataFrame')
for x in stock:
    print(x)
print('\n')

# For Loop - DataFrame, Decomposed
print('For Loop - DataFrame, Decomposed')
for index, inventory in stock.iterrows():
    print(index)
    print(inventory)
print('\n')

# For Loop - DataFrame, Decomposed Further
print('For Loop - DataFrame, Decomposed')
for index, inventory in stock.iterrows():
    print(index, 'has', inventory['apples'], 'apples in stock.')
print('\n')

# Apply Function
print('Apply Function - DataFrame')
characters = {'Character 1':['Monica','Gellar','NY'], 'Character 2':['Rachel','Greene','NY'], 'Character 3':['Phoebe','Buffay','NY']}
characters = pd.DataFrame(characters)
characters.index = ['First Name', 'Last Name', 'State']
characters['Length of Character 1'] = characters['Character 1'].apply(len)
print(characters)
```

# Joining Data with Pandas

## Data Merging Basics

```python
import pandas as pd
wards = pd.read_pickle('datasets/ward.p')
census = pd.read_pickle("datasets/census.p")
licenses = pd.read_pickle('datasets/licenses.p')

# Inner Join: intersection between two tables; default of .merge()
# Outer Join: union of two tables; all items in both tables; default of .merge_ordered()
# Left Join: all items in the left table only, plus those that are in both the left and right tables
# Right Join: all items in the right table only, plus those that are in both the left and right tables

# One-to-One
    # for every row on the left table, there is one row on the right table
    # Number of Rows: same
    # Number of Columns: sum of the number columns both tables minus the number of columns on which the columns merged
wards_census = wards.merge( # left_table.merge(
    census, # right_table
    on='ward', # variable which connects the tables
    suffixes = ('_wards','_census') # add suffix to duplicate column names
)
wards.shape # Result: (50,4)
census.shape # Result: (50,6)
wards_census.shape # Result: (50, 9)

# One-to-Many
    # for every row on the left table, there are one or more rows on the right table
    # Number of Rows: number of rows in right table
    # Number of Columns: sum of the number columns both tables minus the number of columns on which the columns merged
wards_licenses = wards.merge(licenses, on=['ward'], suffixes=('_wards','_licenses'))
wards.shape # Result: (50,4)
licenses.shape # Result: (10000, 6)
wards_licenses.shape # Result: (10000, 9)

# Multiple Tables
wards_census_licenses = wards.merge(census, on='ward', suffixes=('_wards','_census')) \
.merge(licenses, on='ward')
wards_census_licenses.head()
```

## Bootstrap Distributions
Bootstrapping: treat the dataset as a sample and use it to build a theoretical population
- Resample the same size as the original dataset
- Calculate the statistic of interest
- Repeat the above steps many times

Standard error: standard deviation of the statistic of interest


```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
coffee = pd.read_feather('datasets/coffee_ratings_full.feather')

coffee.reset_index(inplace=True)  
coffee_sample = coffee.sample(frac=1, replace=True) # shuffles the order; while the number of items in the sample are the same as the population, some items may appear multiple times and others may be left out
coffee_sample["index"].value_counts() # shows the value counts of each index value
coffee["index"].nunique() - coffee_sample["index"].nunique() # 493 values missing from the sample

# Bootstrap Distribution
mean_flavors = []
for i in range(100):
    mean_flavors.append(np.mean(coffee_sample.sample(frac=1, replace=True)['flavor']))
plt.hist(mean_flavors)
plt.show()

sample_stderr = np.std(mean_flavors, ddof=1) # standard error is the stdev of the statistic of interest
pop_stdev_est = sample_stderr * np.sqrt(100)

# Confidence Interval
coffee_bootstrap_mean = np.mean(mean_flavors) 
coffee_bootstrap_upper1 = coffee_bootstrap_mean + (np.std(mean_flavors, ddof=1) * np.sqrt(100))
coffee_bootstrap_lower1 = coffee_bootstrap_mean - (np.std(mean_flavors, ddof=1) * np.sqrt(100))

# 95% Confidence Interval
    # Option 1
coffee_bootstrap_lower2 = np.quantile(mean_flavors, 0.025)
coffee_bootstrap_upper2 = np.quantile(mean_flavors, 0.975)
    # Option 2
from scipy.stats import norm
coffee_bootstrap_lower3 = norm.ppf(0.025, loc=np.mean(mean_flavors), scale=np.std(mean_flavors, ddof=1))
coffee_bootstrap_upper3 = norm.ppf(0.975, loc=np.mean(mean_flavors), scale=np.std(mean_flavors, ddof=1))
```

# Hypothesis Testing in Python

## Introduction to Hypothesis Testing
- Type 1 Error: false positive (wrongly reject null hypothesis)
- Type 2 Error: false negative (wrongly fail to reject null hypothesis)

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
stack = pd.read_feather('datasets/stack_overflow.feather')

# Z Score - standardized value
hypothesized_mean = 110000  # Hypothesized Mean
sample_mean = stack['converted_comp'].mean() # Sample Mean
bootstrap_mean = [] # Bootstrap 
for i in range(5000): bootstrap_mean.append(np.mean(stack.sample(frac=1, replace=True)['converted_comp']))
std_error = np.std(bootstrap_mean, ddof=1) # Standard Error
z_score = (sample_mean - hypothesized_mean)/std_error # Z Score
p_value = 1 - norm.cdf(z_score, loc=0, scale=1) # P Value
```
